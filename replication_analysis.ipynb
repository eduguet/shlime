```python
names = ['lime only biased classifier', 'lime one uncorrelated column', 'lime two uncorrelated columns', 'shap only biased classifier', 'shap one uncorrelated column', 'shap two uncorrelated columns']
colors = ['#bababb', '#1670a5', '#b54029', '#4bb0d9', '#9467bd', '#8c564b']
desired_feature_order = ['African-American', 'Uncorrelated Feature 1', 'Uncorrelated Feature 2', 'Others']
name_value = 0
row = 0
col = 0
fig, axes = plt.subplots(2, 3, figsize=(15, 6))
for data in total_data:
    # Initialize the figure

    # Extract the keys (group ids) and the values (features) from the dictionary
    categories = list(data.keys())  # Group 1, 2, 3
    features = list({k[0] for v in data.values() for k in v})  # Unique feature names across all groups
    
    # Ensure features are ordered according to the desired order
    ordered_features = [f for f in desired_feature_order if f in features]

    # Create a dictionary to hold the feature values for each group
    group_data = {category: {feature: 0 for feature in features} for category in categories}

    # Fill in the group_data with actual values
    for category, feature_list in data.items():
        for feature, value in feature_list:
            group_data[category][feature] = value

    # Convert group_data into a format suitable for stacked bars
    # Prepare the y values for each feature and each group
    values = {feature: [group_data[category][feature] for category in categories] for feature in features}

    # Define the positions for the bars (x-axis)
    indices = np.arange(len(categories))

    # Plot each feature's values as stacked bars
    bottom = np.zeros(len(categories))  # To stack the bars
    color_index = 0
    for feature, feature_values in values.items():
        axes[row][col].bar(indices, feature_values, bottom=bottom, label=feature, color=colors[color_index])
        color_index = color_index + 1
        bottom += np.array(feature_values)  # Update the bottom to stack the next feature

    # Set the x-axis labels
    axes[row][col].set_xticks(indices)
    axes[row][col].set_xticklabels([str(category) for category in categories])

    # Add title, labels, and legend
    axes[row][col].set_title(names[name_value])
    axes[row][col].set_xlabel('Feature Importance Rank')
    if col == 0:
        axes[row][col].set_ylabel('Occurance')
    if name_value == 2:
        axes[row][col].legend(title="Features", bbox_to_anchor=(1.05, 1), loc='upper left')
    name_value = name_value + 1
    if name_value == 3:
        row = 1
        col = 0
    else:
        col = col + 1

    # Display the plot
plt.tight_layout()
plt.show()
```
